```python
import pandas as pd
df_s = pd.read_csv ('summary.csv')
df_s.head(2)

from plotnine import *
# don't truncate rows and columns when displaying them
pd.set_option('display.max_rows', 999)
pd.set_option('display.max_columns', 999)

```

### DATA ANALYSIS: CONGRESS MEMBERS  

```python
#FITLER CATEGORY FOR CONGRESS MEMBERS  

df_s_clean = df_s[~df_s.CATEGORY.str.contains("TOTALS")]
df_s_clean2 = df_s_clean[df_s_clean['BIOGUIDE_ID'].notnull()]
df_s_clean3 = df_s_clean2[df_s_clean2.AMOUNT > 0] 
df_s_clean3
```

```python
#How much money does the average congressional office spend in 2021 Q2?
#describe the measures of central tendency
#describe the measures of variance after looking at the distribution
#describe any outliers

# Pivot Table congressional members
pd.set_option('display.max_rows', 600)
office = df_s_clean3.pivot_table(index='OFFICE', values='AMOUNT', aggfunc='sum').round(1).fillna('')
office.sort_values(by='AMOUNT')
#office.info()
```

```python

office_new = office.query('AMOUNT < 938756.8')
(
    ggplot(office, aes(x='AMOUNT')) 
        + geom_histogram()
        + theme(figure_size=(16, 8)) # full width
)

```

```python
#How much money does the average congressional office spend in 2021 Q2? 
#mean = 312627.32
#median = 313922.9


office_mean = office["AMOUNT"].mean()
print(office_mean)

office_median = office["AMOUNT"].median()
print(office_median)

print(office['AMOUNT'].std())

```

```python
#measure of variance for office 
#standart deviation = 38791.15

office.AMOUNT.describe().round(1)
```

```python
#calculate mode 
#mode = 281500.8
import statistics

statistics.mode(office.AMOUNT)
```

```python
#outliers 
outliers = office.query('AMOUNT < 100000')

outliers
```
